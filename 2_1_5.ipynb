```python
import numpy as np
import matplotlib.pyplot as plt
from oracles import LogCoshL2Oracle, ExponentialLossL2Oracle, hess_vec_finite_diff
```

```python
np.random.seed(78)

m = 20
n = 5

A = np.random.randn(m, n)
b = np.random.choice([-1.0, 1.0], size=m)
regcoef = 0.1


def matvec_Ax(x):
    return A.dot(x)

def matvec_ATx(x):
    return A.T.dot(x)

def matmat_ATsA(s):
    return A.T.dot(s[:, None] * A)


oracle_logcosh = LogCoshL2Oracle(
    matvec_Ax=matvec_Ax,
    matvec_ATx=matvec_ATx,
    matmat_ATsA=matmat_ATsA,
    b=b,
    regcoef=regcoef
)

oracle_exp = ExponentialLossL2Oracle(
    matvec_Ax=matvec_Ax,
    matvec_ATx=matvec_ATx,
    matmat_ATsA=matmat_ATsA,
    b=b,
    regcoef=regcoef
)


def check_oracle(oracle, name, trials=7):
    print(f"\n{name}")
    print("-" * len(name))

    for k in range(trials):
        x = np.random.randn(n)
        v = np.random.randn(n)

        hv_true = oracle.hess_vec(x, v)
        hv_fd = hess_vec_finite_diff(oracle.func, x, v)

        abs_err = np.linalg.norm(hv_true - hv_fd)
        rel_err = abs_err / max(np.linalg.norm(hv_fd), 1e-12)

        print(f"trial {k+1}: abs_err = {abs_err:.3e}, rel_err = {rel_err:.3e}")


check_oracle(oracle_logcosh, "LogCoshL2Oracle")
check_oracle(oracle_exp, "ExponentialLossL2Oracle")
```
